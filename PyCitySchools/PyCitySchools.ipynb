```python
# Dependencies and Setup
import pandas as pd

import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# Determine the list of unique schools within the database
school_data_complete["school_name"].unique()
```

```python
# Determine the number of unique schools, total number of students and the average scores for maths and reading.
uniqueschoools = school_data_complete['school_name'].unique()
totalschools = len(uniqueschoools)
totalstudents = school_data_complete['Student ID'].count()
totalbudget = school_data['budget'].sum()
avgmath = school_data_complete['maths_score'].mean()
avgread = school_data_complete['reading_score'].mean()
                                   
# Determine the pass rate for each and then determine the overall pass rate (MUST NEED TO PASS BOTH)
mathpass = school_data_complete.query('maths_score >=50')['School ID'].count()/ totalstudents*100
readpass = school_data_complete.query('reading_score >=50')['School ID'].count()/ totalstudents*100     
overall = school_data_complete[(school_data_complete['maths_score'] >= 50) & (school_data_complete['reading_score'] >= 50)]['Student ID'].count()*100/totalstudents
  
# Store all of these in a dataframe    
area_summary = pd.DataFrame({
    "Total Schools": [totalschools],
    "Total Students": [totalstudents],
    "Total Budget": [totalbudget],
    "Average Reading Score": [avgread],
    "Average Math Score": [avgmath],
    "% Passing Reading":[readpass],
    "% Passing Math": [mathpass],
    "Overall Passing Rate": [overall]
})                                   

# Format some of the outputs. 
area_summary= area_summary.style.format({"Total Budget": '{:,}', 
                                      "Total Students": '{:,}',})

area_summary
                                   
                                   
```

```python
# Create a df that will categorise stats based off school names

Listofchools = school_data_complete.groupby(["school_name"])

typeofschool = Listofchools["type"].first()
totalstudentps = Listofchools["Student ID"].count()
totalbudgetps = Listofchools["budget"].first()
avgstudentbudget = totalbudgetps/ totalstudentps
avgmaths = Listofchools["maths_score"].mean()
avgread = Listofchools["reading_score"].mean()
passmathpercent = school_data_complete[school_data_complete["maths_score"] >= 50].groupby(["school_name"])["maths_score"].count()/totalstudentps*100
passreadpercent = school_data_complete[school_data_complete["reading_score"] >= 50].groupby(["school_name"])["reading_score"].count()/totalstudentps*100
overallpassrate = school_data_complete[(school_data_complete['reading_score'] >= 50) & (school_data_complete['maths_score'] >= 50)].groupby('school_name')['Student ID'].count()*100/totalstudentps

# Create a dataframe to store all of the above calculations
per_school_summary = pd.DataFrame({"School Type":typeofschool,
                                  "Total Students":totalstudentps,
                                  "Total School Budget":totalbudgetps,
                                  "Per Student Budget":avgstudentbudget,
                                  "Average Math Score":avgmaths,
                                  "Average Reading Score":avgread,
                                  "% Passing Math":passmathpercent,
                                  "% Passing Reading":passreadpercent,
                                  "Overall Passing Rate":overallpassrate})

# Format some of the outputs into the desired format.
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)




per_school_summary
```

```python
# Grab the top 5 performing schools by overall passing rate.

top_schools = per_school_summary.sort_values(by='Overall Passing Rate', ascending=False).head(5)
top_schools
```

```python
# Now determine the bottom 5 performing schools by overall passing rate.

bottom_schools = per_school_summary.sort_values(by='Overall Passing Rate', ascending=True).head(5)
bottom_schools
```

```python
# For each year level, calculate the average math scores for each school through creating a series.

mgrade9 = school_data_complete.loc[(school_data_complete["year"] == 9)]
mgrade10 = school_data_complete.loc[(school_data_complete["year"] == 10)]
mgrade11 = school_data_complete.loc[(school_data_complete["year"] == 11)]                            
mgrade12 = school_data_complete.loc[(school_data_complete["year"] == 12)]
                                
mgrade9_scores = mgrade9.groupby(["school_name"])["maths_score"].mean()
mgrade10_scores = mgrade10.groupby(["school_name"])["maths_score"].mean()
mgrade11_scores = mgrade11.groupby(["school_name"])["maths_score"].mean()
mgrade12_scores = mgrade12.groupby(["school_name"])["maths_score"].mean()
                                
maths_scores_by_year = pd.DataFrame({"9th": mgrade9_scores, "10th": mgrade10_scores,
                                "11th": mgrade11_scores, "12th": mgrade12_scores})  


maths_scores_by_year
```

```python
# For each year level, calculate the average reading scores for each school through creating a series.

rgrade9 = school_data_complete.loc[(school_data_complete["year"] == 9)]
rgrade10 = school_data_complete.loc[(school_data_complete["year"] == 10)]
rgrade11 = school_data_complete.loc[(school_data_complete["year"] == 11)]                            
rgrade12 = school_data_complete.loc[(school_data_complete["year"] == 12)]
                                
rgrade9_scores = rgrade9.groupby(["school_name"])["reading_score"].mean()
rgrade10_scores = rgrade10.groupby(["school_name"])["reading_score"].mean()
rgrade11_scores = rgrade11.groupby(["school_name"])["reading_score"].mean()
rgrade12_scores = rgrade12.groupby(["school_name"])["reading_score"].mean()
                                
reading_scores_by_year = pd.DataFrame({"9th": rgrade9_scores, "10th": rgrade10_scores,
                                "11th": rgrade11_scores, "12th": rgrade12_scores})  


reading_scores_by_year
```

```python
# create spending bins
spending_bins = [0, 585, 630, 645, 680]
labels = ['< $585', "$585 - 614", "$615 - 644", "$645-680"]
school_data_complete['Spending Ranges (Per Student)'] = pd.cut(school_data_complete['budget']/school_data_complete['size'], spending_bins, labels = labels)

#group by spending
by_spending = school_data_complete.groupby('Spending Ranges (Per Student)')

#calculations
avg_math = by_spending['maths_score'].mean()
avg_read = by_spending['reading_score'].mean()
pass_math = school_data_complete[school_data_complete['maths_score'] >= 50].groupby('Spending Ranges (Per Student)')['Student ID'].count()/by_spending['Student ID'].count()*100
pass_read = school_data_complete[school_data_complete['reading_score'] >= 50].groupby('Spending Ranges (Per Student)')['Student ID'].count()/by_spending['Student ID'].count()*100
overall = school_data_complete[(school_data_complete['reading_score'] >= 50) & (school_data_complete['maths_score'] >= 50)].groupby('Spending Ranges (Per Student)')['Student ID'].count()/by_spending['Student ID'].count()*100

            
# Create a dataframe with the above calculations           
scores_by_spend = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    "% Passing Math": pass_math,
    "% Passing Reading": pass_read,
    "Overall Passing Rate": overall
            
})
            


# Format the results
scores_by_spend= scores_by_spend.style.format({"Average Math Score": '{:.2f}', 
                             "Average Reading Score": '{:.2f}', 
                             "% Passing Math": "{:.2f}",
                             "% Passing Reading": "{:.2f}",                  
                             "Overall Passing Rate": "{:.2f}"})

scores_by_spend
```

```python
# Create size bins
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]


# Use pd.cut to arrange school size into bins
per_school_summary['School Size'] = pd.cut(per_school_summary["Total Students"], size_bins, labels=labels)


# Arrange by school size and calculate the means. 
group_size = per_school_summary.groupby(['School Size'])   
size_summary = group_size.mean()

# Print the results with the given criteria
size_summary[['Average Math Score' ,
                'Average Reading Score',
                '% Passing Math',
                '% Passing Reading',
                'Overall Passing Rate']]
```

```python
# Arrange schools by school type
school_type = per_school_summary.groupby(['School Type'])              
# Calculate values through the type of school
type_summary = school_type.mean()

# Print the result using the same criteria as above
type_summary[['Average Math Score',
                'Average Reading Score',
                '% Passing Math',
                '% Passing Reading',
                'Overall Passing Rate']]
```
